### (heart) Heart disease
For the purpose of this analysis, we consider all codes in the ICD-10 Chapter IX: Diseases of the circulatory system that are related to the heart, but not primarily to blood vessels. These codes are:
- I05-I09
- I11
- I20-I25 (added after comparing data with validation data from 2 websites)
- I26-I28
- I30-I52

### Cancer
- (cancer_all): C00-D48 (Neoplasms)
- (cancer_lung) Lung and bronchial cancer: C34 (and sub)
- (cancer_anal) Colon and rectal cancer: C18-C21
- (cancer_breast) (Female) breast cancer: C50

### (loresp) Chronic lower respiratory disease
E.g., Bronchitis (J40), Emphysema (J43), Asthma (J45)
- J40-47

### (accident) Accidents (unintentional injuries)
Personally, I would also consider "Complications of medical and surgical care" as accicents, but we will stick with ICD10 conventions here who have three chapters reserved for accidents.
- V01-X59

### (stroke) Stroke
- I61-64

### (alzheimer) Alzheimer's
- F00
- G30 (added after comparing data with validation data from 2 websites)

### (diabetes) Diabetes
- E10-14

### (influpneu) Influenza and pneumonia
- J09-18

### (kidney) Kidney disease (somewhat more broadly, Diseases of the genitourinary system: urinary system)
- N00-N29

### (suicide) Suicide
- X60-84

### Homicide
- (homicide_all) All: X85–Y09
- (homicide_terror) Terrorism: Y38 (potentially overlapping with next)
- (homicide_firearm) Assault involving firearms: X93-X95
- (homicide_military) Military operations: Y37 (potentially overlapping with next)
- (homicide_legmil) Legal intervention and operations of war: Y35-36

### Terrorism
(After comparing our results with available lists in Top10 causes on the Internet, we realized that Y38 is actually never used in CDC. They use U01 instead, so let's move it to its own category:)
- (terrorism) U-01

### Overdoses
- (overdose) T36-T50

```python
def get_cdc_icd_blocks(icd_block_range_str):
    """
    Returns sorted array of ICD-10 block names (first three chars, e.g. "I05")
    based on occurrences in CDC_cause.
    
        icd_block_range_str:    String containing a description of ICD-10 block name ranges,
                                using a lower and an upper boundary, separated by colons ("-")
    
    Example: "I05-I09" will return ["I05", "I06", ..., "I09"]
    """
    
    def get_icd_block_range(start_block, end_block):
        """
        Returns hypothetical (need not exist in cdc_cause) ICD-10 block range
        """
        
        def get_char_code(char):
            """
            "a" -> "00"
            "b" -> "01"
            ...
            """

            if not len(char) == 1:
                raise ValueError("char must be single letter")

            char_pos = np.where(np.array(list(string.ascii_lowercase)) == char.lower())[0][0]
            char_pos = "{:02d}".format(char_pos)
            return(char_pos)

        def get_codenum(block):
            """
            'A00' -> 0
            'A01' -> 1
            ...
            'A99' -> 99
            'B00' -> 100
            ...
            """
            
            return(int(get_char_code(block[0]) + block[1:3]))

        def get_blocks(codenums):
            """
            E.g., [0, 1, 101] -> ['A00', 'A01', 'B01']
            """
            
            def get_block(codenum):
                char   = list(string.ascii_lowercase)[int(np.floor(codenum/100))].upper()
                digits = "{:02d}".format(int(str(codenum)[-2:]))
                block  = char+digits
                return(block)

            return(list(map(get_block, codenums)))

        start_codenum = get_codenum(start_block)
        end_codenum = get_codenum(end_block)

        codenum_range = np.arange(start_codenum, end_codenum+1)

        block_range = get_blocks(codenum_range)

        return(block_range)
    
    # List with sorted unique Blocks (e.g., "I10") in CDC_cause
    all_cdc_blocks = np.array(cdc_cause['Code'].str.slice(0, 3).unique())
    all_cdc_blocks.sort()

    boundaries = np.array(icd_block_range_str.split("-"))
    if len(boundaries) == 1:
        start_block  = boundaries[0]
        end_block    = start_block
    elif len(boundaries) == 2:
        start_block  = boundaries[0]
        end_block    = boundaries[1]
    
    if(len(start_block) > 3 | len(end_block) > 3):
        print("Length of boundary start or end > 3, warning: information might be truncated")
        start_block  = start_block[:3]
        end_block    = end_block[:3]
    elif(len(start_block) < 3 | len(end_block) < 3):
        raise ValueError("Length of boundary start or end < 3, you might be trying to use this function in a way that hasn't been implemented yet")
    
    hypothetical_icd_block_range = get_icd_block_range(start_block, end_block)
    return(all_cdc_blocks[np.isin(all_cdc_blocks, hypothetical_icd_block_range)])
```

```python
def get_mortality_by_icd_block_range(icd_block_range_str, years = None, called_internally = False):
    """
    Returns number of deaths in cdc_cause
    First output is raw sum, second is sum of #death rel to population (only if only one year)

        icd_block_range_str:    String containing a description of ICD-10 block name ranges,
                                using a lower and an upper boundary, separated by colons ("-").
        year:                   Array of years to filter cdc_cause.
        called_internally:      Can be set true to avoid multiple warning messages
                                related to omission of relative death numbers when several years
                                are indicated

    """
    
    block_range = icd_block_range_str
    years = 2009

    if isinstance(years, int):
        years = [years]
    elif years == 'all':
        years = list(cdc_cause['Year'].unique())

    block_list  = get_cdc_icd_blocks(block_range)
    
    cdc_block_and_year_match_idx = np.logical_and(np.isin(cdc_cause['Code'].str.slice(0, 3), block_list),
                                                  np.isin(cdc_cause['Year'], years))
    
    print(cdc_cause.loc[cdc_block_and_year_match_idx, :])
    
    cdc_matches_death_pop = cdc_cause.loc[cdc_block_and_year_match_idx, ['Deaths', 'Population']]
    cdc_matches_death_pop['Relative_Deaths'] = cdc_matches_death_pop['Deaths'] / cdc_matches_death_pop['Population']
    cdc_matches_results = cdc_matches_death_pop.sum(axis=0)[['Deaths', 'Relative_Deaths']]
    return(list(cdc_matches_results))
```

```python
def get_mortality_by_category(cat_codename_icd_dict, years = "all"):
    """
    Returns number of deaths per category as a df.
    First output is raw sum, second is sum of #death rel to population (only if only one year).

        cat_codename_icd_dict:      dict with category codenames as keys and ICD-10 block range descriptions as values.
        years:                      Array of years to filter cdc_cause.

    """

    categories = []
    raw_mortality = []
    rel_mortality = []
    for codename, icd_block_range_str_arr in cat_codename_icd_dict.items():
        cat_sum_raw = 0
        cat_sum_rel_pop = 0

        for icd_block_range_str in icd_block_range_str_arr:
            mortality = get_mortality_by_icd_block_range(icd_block_range_str, years, called_internally = True)
            mortality_raw = mortality[0]
            mortality_rel_pop = mortality[1]
            cat_sum_raw += mortality_raw
            if isinstance(years, int) or len(years) == 1:
                cat_sum_rel_pop += mortality_rel_pop

        if not (isinstance(years, int) or len(years) == 1):
            cat_sum_rel_pop = None

        categories.append(codename)
        raw_mortality.append(cat_sum_raw)
        rel_mortality.append(cat_sum_rel_pop)

    if not (isinstance(years, int) or len(years) == 1):
        print("!! Not returning relative sums, since relative sums don't make sense across more than one year")
    
    df = pd.DataFrame(np.hstack((np.matrix(raw_mortality).T,
                                 np.matrix(rel_mortality).T)),
                      index = categories,
                      columns = ['raw_mortality', 'rel_mortality'])
    
    return(df)
```

```python
def compare_top_causes_2016(our_data_df):
    # Comparison data
    comp_top10 = pd.read_csv('./data/medicalnewstoday_statista_top_deaths_2016.csv')
    comp_top10 = comp_top10.set_index('cause')
    comp_top10[comp_top10 == 0] = None

    # Data for normalization, extracted from cdc:
    pop_2016      = 318810586
    n_deaths_2016 = 2744248

    # Normalize data by number of deaths in 2016
    comp_top10['mnt_ratio'] = comp_top10['mnt_deaths']/n_deaths_2016
    comp_top10['statista_ratio'] = comp_top10['statista_deaths']/n_deaths_2016
    comp_top10['cdc_ratio'] = our_data_df['raw_mortality']/n_deaths_2016

    comp_top10.loc[:, ['cdc_ratio', 'mnt_ratio', 'statista_ratio']].plot.bar()
```
